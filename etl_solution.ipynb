### Loading data into database

#### Create function to connect to database

```python
def get_db_connection():
    connection = psycopg2.connect(
        host = 'localhost',
        database = 'zuma_realtor',
        user ='postgres',
        password ='password',
        port = 5432
    )
    return connection

# Connect to the database
conn = get_db_connection()
```

#### Create function that sets up schema and tables

```python
def create_tables():
    conn = conn = get_db_connection()
    cursor = conn.cursor()
    create_tables_query =''' 
      
            CREATE SCHEMA IF NOT EXISTS zuma_realtors;

            DROP TABLE IF EXISTS zuma_realtors.location_dim CASCADE;
            DROP TABLE IF EXISTS zuma_realtors.sales_dim CASCADE;
            DROP TABLE IF EXISTS zuma_realtors.features_dim CASCADE;
            DROP TABLE IF EXISTS zuma_realtors.property_fact_table CASCADE;

            CREATE TABLE zuma_realtors.location_dim(
                        location_ID SERIAL PRIMARY KEY,
                        addressLine1 VARCHAR(1000),
                        addressLine2 VARCHAR(1000),
                        city VARCHAR(200),
                        state VARCHAR(200),
                        zipCode INT,
                        formattedAddress VARCHAR(300),
                        longitude FLOAT,
                        latitude FLOAT,
                        county VARCHAR(200)
            );

            CREATE TABLE zuma_realtors.sales_dim(
                        sales_ID SERIAL PRIMARY KEY,
                        lastSaleDate VARCHAR(200),
                        lastSalePrice VARCHAR(200),    
            );
  
            CREATE TABLE zuma_realtors.features_dim(
                        feaures_ID SERIAL PRIMARY KEY,
                        bedrooms FLOAT,
                        squareFootage VARCHAR(200),
                        bathrooms FLOAT,
                        lotSize FLOAT,
                        features VARCHAR(200)
   
            );

            CREATE TABLE zuma_realtors.property_fact_table(
                       id SERIAL PRIMARY KEY,
                       feaures_ID INT,
                       location_ID INT,
                       sales_ID INT,
                       assessorID VARCHAR(200),
                       legalDescription VARCHAR(1000),
                       ownerOccupied VARCHAR(1000),
                       subdivision VARCHAR(200),
                       yearBuilt FLOAT,
                       propertyType VARCHAR(200),
                       taxAssessment VARCHAR(1000),
                       owner VARCHAR(1000),
                       propertyTaxes VARCHAR(1000),
                       zoning VARCHAR(1000),
                       FOREIGN KEY(feaures_ID) REFERENCES zuma_realtors.features_dim(feaures_ID),
                       FOREIGN KEY(sales_ID) REFERENCES zuma_realtors.sales_dim(sales_ID),
                       FOREIGN KEY(location_ID) REFERENCES zuma_realtors.location_dim(location_ID)

            );
              '''
    cursor.execute(create_tables_query)
    conn.commit()
    cursor.close()
    conn.close()
```
